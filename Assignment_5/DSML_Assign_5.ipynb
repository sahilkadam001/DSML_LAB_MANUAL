```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
myfile = pd.read_csv("/content/Mall_Customers.csv")
myfile.head()
```

```python
x = myfile.Age
y=myfile.iloc[:,3].values
plt.scatter(x,y)
plt.title("Age vs Annual Income(k $)")
plt.xlabel("Age")
plt.ylabel("Annual Income (k $)")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, marker ="o", c = 'hotpink')
plt.title("Age vs Annual Income(k $)")
plt.xlabel("Age")
plt.ylabel("Annual Income (k $)")
plt.show()
```

```python
z = myfile.iloc[:,4].values
plt.bar(x,z)
plt.title("Age vs spending score")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
sns.boxplot(x =myfile.Age ,y=myfile.iloc[:,3].values,data = myfile)
plt.title("Age vs Annual Income(k $)")
plt.xlabel("Age")
plt.ylabel("Annual Income (k $)")
plt.show()
```

```python
plt.hist(y, edgecolor='black' )
plt.title("Histogram of Annual Income vs No. of people")
plt.xlabel(" No. of People")
plt.ylabel("Annual Income(k $)")
plt.show()
```

```python
mylabel = myfile.CustomerID
plt.pie(y)
plt.show()
```

```python
plt.stackplot(x,y,colors=["b"])
plt.xlabel(" Age")
plt.ylabel("Annual Income(k $)")
plt.show()
```

```python
sns.kdeplot(y)
plt.show()
```

```python
sns.violinplot(x = myfile.Age, y=myfile.iloc[:,3].values, data = myfile)
plt.show()
```
